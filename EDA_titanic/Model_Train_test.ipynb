## Handling missing Numerical values

```python
a = [X_train, X_test, y_train, y_test]

num = ['age', 'sibsp', 'parch', 'fare']
cat = ['pclass', 'sex', 'embarked', 'class', 'who', 'adult_male', 'embark_town', 'alone']
X_train.isna().sum()
```

### Filling missing value using the Simple Imputer

```python
from sklearn.impute import SimpleImputer

SI1= SimpleImputer(strategy="median")
SI2= SimpleImputer(strategy="most_frequent")

X_train['age']= SI1.fit_transform(X_train[['age']])
X_test['age']= SI1.transform(X_test[['age']])

X_train[cat]= SI2.fit_transform(X_train[cat])
X_test[cat]= SI2.transform(X_test[cat])
```
