```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'insurance.csv'
ins = pd.read_csv(path)
ins.head()
```

```python
ins.info()
```

```python
ins.columns
```

```python
ins.describe()
```

```python
ins.isna().sum()
```

```python
num = ['age', 'bmi', 'children', 'charges']
cat = ['sex', 'smoker', 'region']
```

```python
print(f'Numerical Columns in Insurance :{num}')
print(f'Categorial Columns in Insurance :{cat}')
```

# Univariate Analysis

```python
for i in num:
    plt.figure(figsize=(10, 5))
    sns.histplot(
          data=ins[i],
          kde=True)
    plt.show()
```

```python
for i in num:
    plt.figure(figsize=(10,5))
    sns.boxplot(
        data=ins[i],
        orient='h')
    plt.title(f'Box Plot of {i}')
    plt.show()
```

## HeatMap

```python
corr = ins[num].corr()
sns.heatmap(
    data= corr,
    annot=True,
    cmap='Purples_r'
)
```

## Pairplot

```python
sns.pairplot(
    data= ins[num]
)
plt.show()
```
